```python
chunksize = 10_000 
first_chunk = True

if not input_path.is_file():
    raise FileNotFoundError(f"Fichier introuvable : {input_path}")

for i, chunk in enumerate(pd.read_csv(input_path, chunksize=chunksize)):
    chunk.drop(columns=cols_to_drop, errors="ignore", inplace=True)
    mode = "w" if first_chunk else "a"
    header = first_chunk  
    chunk.to_csv(output_path, index=False, mode=mode, header=header)
    
    if first_chunk:
        print("Nombre de colonnes après suppression :", len(chunk.columns))
        first_chunk = False

    print(f"Chunk {i+1} traité.")

print("Fichier nettoyé écrit dans :", output_path)
```
